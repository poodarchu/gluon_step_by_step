```python
# Lazy Evaluation and Automatic Parallelization
```

```python
import time
def do(x, n):
    return [mx.nd.dot(x, x) for i in range(n)]
def wait(x):
    for i in x:
        i.wait_to_read()
```

```python
tic = time.time()
a = mx.nd.ones((1000, 1000))
b = do(a, 50)
print('time for all computations are pushed into backend engin:\n %f sec' % (time.time()-tic))
wait(b)
print('time for all computations are finished:\n %f sec' %(time.time()-tic))
```

```python
n = 10
a = mx.nd.ones((6000,6000))
b = mx.nd.ones((6000,6000), gpu_device)
tic = time.time()
c = do(a, n)
wait(c)
print('Time to finish the CPU workload: %f sec' % (time.time() - tic))
d = do(b, n)
wait(d)
print('Time to finish both CPU/GPU workloads: %f sec' % (time.time() - tic))
```

```python
# If we issue all workloads at the same time, the backend engine will try to parallel the CPU and GPU computations.
tic = time.time()
c = do(a, n)
d = do(b, n)
wait(c)
wait(d)
print('time to finish both CPU and GPU computations: %f sec' %(time.time() - tic))
```
